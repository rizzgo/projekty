```python
import os
import numpy as np
import matplotlib.pyplot as plt

class Stats:
    def __init__(self) -> None:
        self.fitness_history = []
        self.network = None
        self.fitness = None

path="stats"
if not os.path.exists(path):
    os.mkdir(path)
```

```python
p8_16 = np.load("drive/MyDrive/VUT/EVO/pow-50000iters,500tests/run8/runs.npy", allow_pickle=True)
g99_16 = np.load("drive/MyDrive/VUT/EVO/geo-50000iters,500tests/run99/runs.npy", allow_pickle=True)
p8_8 = np.load("drive/MyDrive/VUT/EVO/size8/pow-50000iters,500tests/run8/runs.npy", allow_pickle=True)
g99_8 = np.load("drive/MyDrive/VUT/EVO/size8/geo-50000iters,500tests/run99/runs.npy", allow_pickle=True)
p8_12 = np.load("drive/MyDrive/VUT/EVO/size12/pow-50000iters,500tests/run8/runs.npy", allow_pickle=True)
g99_12 = np.load("drive/MyDrive/VUT/EVO/size12/geo-50000iters,500tests/run99/runs.npy", allow_pickle=True)
```

```python
def fitness_history_graph(runs, path, max_iterations):
    fitness_history_runs = np.array([run.fitness_history for run in runs])

    mins = np.min(fitness_history_runs, axis=0)
    meds = np.median(fitness_history_runs, axis=0)
    maxs = np.max(fitness_history_runs, axis=0)
    var = np.var(fitness_history_runs, axis=0)

    fig = plt.figure(figsize=(12.8, 7.2), dpi=100)
    x = np.arange(1,max_iterations+1)
    plt.rc('font', size=16)
    plt.plot(x,meds, label="median")
    plt.plot(x,mins, label="minimum")
    plt.fill_between(x,mins,maxs,alpha=0.4)
    # plt.fill_between(x,meds-var,meds+var,alpha=0.2)
    plt.legend()
    plt.xlabel("počet iterácií")
    plt.ylabel("fitness")
    fig.savefig(os.path.join(path, "fitness_history_best_16.pdf"), format="pdf")

fitness_history_graph(p8_16, path, 50000)
```
