```python
def boxplot_graph_pdf(runs_to_compare, storage_path):
  fig = plt.figure(figsize=(17.8, 7.2), dpi=100)
  plt.boxplot([np.array([run.fitness for run in runs]) for runs in runs_to_compare], notch=True)
  plt.xticks(np.arange(1, len(runs_to_compare)+1), ["0.01", "0.05", "0.1", "0.4", "0.6", "0.8", "0.9", "0.95", "0.99"])
  plt.rc('font', size=16)
  plt.xlabel("\u03B1")
  plt.ylabel("fitness")
  plt.savefig(os.path.join(storage_path, "boxplot_pow.pdf"), format="pdf")

boxplot_graph_pdf([p01, p05, p1, p4, p6, p8, p9, p95, p99], path)
!cp -R stats drive/MyDrive/VUT/EVO/temp-boxes/box
```
